```python
# Lowercase
df_missing['Employee'] = df_missing['Employee'].str.lower()

# Uppercase
df_missing['Region'] = df_missing['Region'].str.upper()
```

```python
df_missing[df_missing['Product'].str.contains('phone', case=False)]
```

```python
df_missing['Product'] = df_missing['Product'].str.replace('Smartphone', 'Mobile', regex=False)
```

```python
df_missing['Employee'] = df_missing['Employee'].str.replace('[^a-zA-Z ]', '', regex=True)
```

```python
df_missing['Region_Code'] = df_missing['Region'].str.slice(0, 3)
```

```python
df_missing['Employee'] = df_missing['Employee'].str.strip()
```

```python
df_missing['Employee_Product'] = df_missing['Employee'].str.title() + ' - ' + df_missing['Product']
```

```python
import pandas as pd

df_missing['Month_Num'] = pd.to_datetime(df_missing['Month'], format='%B').dt.month
```

```python
df_missing[['Month', 'Month_Num']].head()
```

```python
df_missing['Month'] = pd.to_datetime(df_missing['Month'], errors='coerce')

df_missing['Month'].head()
df_missing['Month'].dtype
```

```python
df_missing['Year'] = df_missing['Month'].dt.year
df_missing['Month_Num'] = df_missing['Month'].dt.month
df_missing['Day_Name'] = df_missing['Month'].dt.day_name()
```

```python
monthly_sales = df_missing.groupby('Month_Num')['Sales'].sum().sort_index()
print(monthly_sales)
```

```python
monthly_sales.plot(kind='bar', title='Monthly Sales Trend')
```

```python
best_month = df_missing.groupby('Month_Num')['Profit'].sum().idxmax()
print(f"Best performing month (by total profit): {best_month}")
```

```python
weekday_profit = df_missing.groupby('Day_Name')['Profit'].mean().sort_values(ascending=False)
print(weekday_profit)
```

```python
df_missing['Profit_Percent'] = df_missing.apply(
    lambda row: (row['Profit'] / row['Sales']) * 100 if row['Sales'] != 0 else 0,
    axis=1
)
df_missing[['Sales', 'Profit', 'Profit_Percent']].head()
```

```python
df_missing['Region'] = df_missing['Region'].map(lambda x: x.upper())
df_missing['Region'].head()
```

```python
region_map = {'North': 'N', 'South': 'S', 'East': 'E', 'West': 'W'}
df_missing['Region'] = df_missing['Region'].replace(region_map)
```

```python
df_missing['Product'] = df_missing['Product'].replace({
    'laptop': 'Laptop',
    'LAPTOP': 'Laptop',
    'mobile': 'Mobile'
})
```

```python
df_missing['Profit'] = df_missing['Profit'].replace(-999, np.nan)
```

```python
df_missing = df_missing.rename(columns={
    'Sales': 'Total_Sales',
    'Profit': 'Net_Profit'
})
df_missing.head()
```

```python
df_missing = df_missing.drop('Profit Margin', axis=1)
```

```python
df_missing = df_missing.drop(['Month_Num', 'Year'], axis=1)
```
